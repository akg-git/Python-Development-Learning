# MatplotLib library**

Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python
It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits

Key features:
* Create 2D plots (line, scatter, bar, histogram, etc.)
* Support for multiple plot types and customization options
* Integration with NumPy and Pandas
* Publication-quality figures

Basic usage example:
* Create a figure and axes
* Plot data using various methods (plot, scatter, bar, etc.)
* Customize (labels, titles, legends, colors, etc.)
* Display using plt.show()

```python
import matplotlib.pyplot as plt

import numpy as np
```

```python
xaxis = np.arange(1,30)
yaxis = np.arange(1,30)
```

```python
# create a scatter plot of squared values
plt.scatter(xaxis, yaxis**2)
plt.xlabel('X Axis Label') 
plt.ylabel('Y Axis Label')
plt.title('2D Scatter Plot Example')

#plt.savefig('2d_scatter_plot.png')  # Save the plot as a PNG file
```

```python
plt.plot(xaxis, yaxis**3)
plt.xlabel('X Axis Label') 
plt.ylabel('Y Axis Label')
plt.title('2D Scatter Plot Example')
```

```python
plt.plot(xaxis*2, (yaxis/3)+5, color='g', linestyle='--', marker='o', markerfacecolor='b', markersize=5)
plt.xlabel('X Axis Label') 
plt.ylabel('Y Axis Label')
plt.title('2D Scatter Plot: 2x+Y/3+5=0')
```
