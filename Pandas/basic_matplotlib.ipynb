# MatplotLib library**

Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python
It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits

Key features:
* Create 2D plots (line, scatter, bar, histogram, etc.)
* Support for multiple plot types and customization options
* Integration with NumPy and Pandas
* Publication-quality figures

Basic usage example:
* Create a figure and axes
* Plot data using various methods (plot, scatter, bar, etc.)
* Customize (labels, titles, legends, colors, etc.)
* Display using plt.show()

```python
import matplotlib.pyplot as plt

import numpy as np
```

```python
xaxis = np.arange(1,30)
yaxis = np.arange(1,30)
```

```python
# create a scatter plot of squared values
plt.scatter(xaxis, yaxis**2)
plt.xlabel('X Axis Label') 
plt.ylabel('Y Axis Label')
plt.title('2D Scatter Plot Example')

#plt.savefig('2d_scatter_plot.png')  # Save the plot as a PNG file
```

```python
plt.plot(xaxis, yaxis**3)
plt.xlabel('X Axis Label') 
plt.ylabel('Y Axis Label')
plt.title('2D Scatter Plot Example')
```

```python
plt.plot(xaxis*2, (yaxis/3)+5, color='g', linestyle='--', marker='o', markerfacecolor='b', markersize=5)
plt.xlabel('X Axis Label') 
plt.ylabel('Y Axis Label')
plt.title('2D Scatter Plot: 2x+Y/3+5=0')
```

```python
#Lets discuss about sin and cosine graph

#define the range of x values from -4 to 4 with step of 1000 unifrorm points
x = np.linspace(-4 * np.pi, 4 * np.pi, 1000)

# Sine Graph
plt.title('Sine Functions')
#calculate the corresponding y values for sine function within the range of [-1, 1]
ysin = np.sin(x)
#plot the sine wave
plt.plot(x, ysin, label='Sine Wave', color='b', linestyle='--')

#Customize x-axis ticks to show multiples of 
pi_ticks = np.arange(-4*np.pi, 5*np.pi, np.pi)  # from -4 to +4
pi_labels = [f"{i}" if i != 0 else "0" for i in range(-4, 5)]
plt.xticks(pi_ticks, pi_labels)

# Add labels and show the plot
plt.xlabel(r"$\ X\ values (\pi\ radians) $")
plt.ylabel('Y values (numeric)')
plt.show()

# Cosine Graph
plt.title('Cosine Functions')
# calculate the corresponding y values for cosine function within the range of [-1, 1]
ycos = np.cos(x)
# plot the cosine wave
plt.plot(x, ycos, label='Cosine Wave', color='g', linestyle='--')
# Customize x-axis ticks to show multiples of 

#reuse customized pi_ticks and pi_labels from sine graph
plt.xticks(pi_ticks, pi_labels)

# Add labels and show the plot
plt.xlabel(r"$\ X\ values (\pi\ radians) $")
plt.ylabel('Y values (numeric)')
plt.show()
```

```python
#Plot a Pie Chart

labels = [ 'CS', 'ECE', 'ME', 'CE', 'EE']
sizes = [270, 70, 45, 28, 182]
colors = ['lightskyblue', 'yellowgreen', 'lightcoral', 'lightpink', 'lightgreen']
# explode = (0.1, 0, 0, 0, 0)  # explode the 1st slice (CS)

plt.title('Departmental Student Distribution')
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
