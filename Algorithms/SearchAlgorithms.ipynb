**Binary Search**

```python
def binary_search(lst, key):
    low = 0
    high = len(lst) - 1

    while low <= high:
        mid = (low + high)  // 2
        if (key == lst[mid]):
            return mid
        elif (key > lst[mid]):
            low = (mid + 1)
        else:
            high = (mid - 1)
    return -1
```

```python
array = [17, 74, 71, 31, 81, 64, 66, 28, 50, 77, 67, 59, 63, 88, 92, 95, 66, 31, 93, 31]
sorted_array = sorted(array)
#Sorted array is :  [17, 28, 31, 31, 31, 50, 59, 63, 64, 66, 66, 67, 71, 74, 77, 81, 88, 92, 93, 95]
key = int(input('Enter the element to be searched: '))
result = binary_search(sorted_array, key)
if result != -1:
    print("{} found at {} index".format(key, result))
else:
    print("Search completed. Alas!! {} not found.".format(key))
```
