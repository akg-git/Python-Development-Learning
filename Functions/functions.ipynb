#Example1: Temperature Conversion

Conversion of temperature from celcius to ferhenhit and ferhenhit to celcius

```python
def convert_temperature(temp, unit):
    '''
    Docstring for convert_temperature
    
    :param temp: Description
    :param unit: Description
    '''

    # Conversion from celcius to ferhenhit
    if unit == 'C':
        if temp < -273.15:
            raise ValueError("Temperature below absolute zero is not possible")
        return (temp * (9/5)) + 32
    # Conversion from  ferhenhit to celcius
    elif unit == 'F':
        if temp < -459.67:
            raise ValueError("Temperature below absolute zero is not possible")
        return (temp - 32) * (5/9)
    else:
        raise ValueError("Unit must be 'C' for Celsius or 'F' for Fahrenheit")
```

```python
temp = 198
unit = 'F'
converted_temp = convert_temperature(temp, unit)

# if unit == 'C':
#    print(f"{temp} C = {round(converted_temp,2)} F")
# elif unit == 'F':
#    print(f"{temp} F = {round(converted_temp,2)} C")
# else:
#     print("Invalid unit provided. Enter 'C' for Celsius or 'F' for Fahrenheit.")

# string comprehension
print(f"{temp} °{unit} = {round(converted_temp,2)} {'°F' if unit == 'C' else '°C' if unit =='F' else 'Invalid Unit' }")
```

#Example2: Password Validator

It will check and validate the given password

```python
def validate_password(password):
    '''
    Docstring for validate_password
    
    :param password: Description
    '''

    if len(password) < 8:
        return False
    if not any(char.isdigit() for char in password):
        return False
    elif (password[0].isdigit() or password[-1].isdigit()):
        return False
    if not any(char.isupper() for char in password):
        return False
    if not any(char.islower() for char in password):
        return False
    if not any(char in '!@#$%^&*()-_=+[]{}|;:,.<>?/' for char in password):
        return False
    return True

input_password = "StrongPass1!"
is_valid = validate_password(input_password)
print(f"Is the password '{input_password}' valid? => {is_valid}")
print(validate_password("weakpass")) 
print(validate_password("4Short1!"))
print(validate_password("$Short0!"))
```

#Example3: Calculate total Cost of the products present in the cart

#sample cart data
{
    {'name':'Apple', 'price':30, quantity: 4}
    {'name':'Orange', 'price':60, quantity: 3}
    {'name':'Banana', 'price':15, quantity: 10}
}

```python
def calculate_cart_total(cart):
    '''
    Docstring for calculate_cart_total
    
    :param cart: Description
    '''
    total_price = 0
    for item in cart:
        total_price += item['price'] * item['quantity']
    return total_price


#sample cart data
cart = [
    {'name':'Apple', 'price':30, 'quantity': 4},
    {'name':'Orange', 'price':60, 'quantity': 3},
    {'name':'Banana', 'price':15, 'quantity': 10}
]

print(f"Your required stuffs are just {calculate_cart_total(cart)} Rs/-")
```

#Example 5: Word Frequency from an external file 

```python

def word_frequency(file_path):
    '''
    Docstring for word_frequency
    
    :param file_path: Description
    '''
    frequency = {}
    with open(file_path, 'r') as file:
        for line in file:
            words = line.split()
            for word in words:
                word = word.lower().strip('.,!?;"\'()[]{}') # strip used to Remove punctuation
                #main logic to count word frequency
                frequency[word] = frequency.get(word, 0) + 1
        return frequency

filepath = 'sample.txt'
print(word_frequency(filepath))
```
