Lambda Function:

Lambda functions are small anonymous functions defined using the lambda expressions using lambda keyword.

They can have any number of arguments but only one expression.

```python
#addition operation
addition = lambda x, y, z, w: x + y + z + w
type(addition)
print(addition(10, -2, 13, 24))
```

Check if given number is Odd or not

```python
# odd value out
odd = lambda num: num%2==1
print(odd(7))
print(odd(9))
print(odd(90))
```

Even-Odd function using lamda expression alongside List comprehension

```python
numlist = [17, 74, 71, 31, 81, 64, 66, 28, 50, 77, 67, 59, 63, 88, 92, 95, 66, 31, 93, 31]

#odd numbers from numlist
odd_numbers = list(filter(lambda num: num%2 ==1, numlist))

#even numbers from numlist
even_numbers = list(filter(lambda num: num%2 ==0, numlist))

#classify using list comprehension
even_odd_classifier = ["ODD" if num in odd_numbers else "EVEN" for num in numlist]
print(even_odd_classifier)
```

Else-if ladder in the list comprehension

```python
numlist = [17, 54, 71, 31, 81, 64, 66, 28, 50, 77, 67, 59, 60, 88, 92, 95, 66, 31, 93, 31]

labels = [
    "divisible by both 2 and 3" if n % 2 == 0 and n % 3 == 0
    else "divisible by 2" if n % 2 == 0
    else "divisible by 3" if n % 3 == 0
    else "other"
    for n in numlist
]

print(labels)
```
